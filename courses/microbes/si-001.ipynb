```python
import matplotlib.pyplot as plt
import skimage.color as color

from skimage import data
from skimage.exposure import histogram
from skimage import io
```

```python
image = io.imread("scene00701.png")
plt.imshow(image)
```

```python
from skimage.feature import canny
image_gray = color.rgb2gray(image)
```

```python
plt.imshow(image_gray, cmap=plt.cm.gray)
```

```python
from skimage import data
from skimage.exposure import histogram

hist, hist_centers = histogram(image_gray)

fig, axes = plt.subplots(1, 2, figsize=(8,3))
axes[0].imshow(image_gray, cmap=plt.cm.gray)
axes[0].axis('off')
axes[1].plot(hist_centers, hist, lw=2)
axes[1].set_title("histogram")
```

```python
edges = canny(image_gray)
```

```python
plt.imshow(edges, cmap=plt.cm.gray)
```

```python
from scipy import ndimage as ndi

fill_image = ndi.binary_fill_holes(edges)
plt.imshow(fill_image, cmap=plt.cm.gray)
```
