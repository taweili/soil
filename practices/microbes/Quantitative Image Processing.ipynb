```python
# Our standard modules
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Utilities for image processing
import skimage.io
import skimage.exposure
import skimage.measure
import skimage.color as color

# To allow for inline rendering of plots.
%matplotlib inline
```

```python
microbes_img = skimage.io.imread("scene00701.png")
```

```python
plt.imshow(microbes_img)
```

```python
microbes_img_gray = color.rgb2gray(microbes_img)
```

```python
plt.imshow(microbes_img_gray, cmap=plt.cm.Greys_r)
```

```python
hist, bins = skimage.exposure.histogram(microbes_img_gray)

plt.plot(bins, hist, linewidth=1)
```

```python
treshold_img = microbes_img_gray < 0.6

plt.imshow(treshold_img, cmap=plt.cm.gray)
```
